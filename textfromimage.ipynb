This project essentially involves using Gemini to generate text from image input. To complete this, I imported the necessary libraries and utilized the GOOGLE_API_KEY. I employed a model called 'gemini-pro-vision,' and with the help of 'to_markdown,' I generated text based on the given input image

```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown
```

```python
def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
# Used to securely store your API key
from google.colab import userdata

from google.colab import userdata
GOOGLE_API_KEY = userdata.get('gemini')
genai.configure(api_key=GOOGLE_API_KEY)
```

# Generate text from image
Gemini provides a multimodal model (gemini-pro-vision) that accepts both text and images and inputs. The GenerativeModel.generate_content API is designed to handle multimodal prompts and returns a text output.

```python
# use the gemini-pro-vision model and pass the image to the model with generate_content.
model = genai.GenerativeModel('gemini-pro-vision')
```

```python
from PIL import Image
import IPython.display as display
```

```python
# Open the image
img = Image.open('/content/drive/MyDrive/gemini/img gemini/dog.jpeg')

# Display the resized image
display.display(img)
```

```python
response = model.generate_content(img)
to_markdown(response.text)
```

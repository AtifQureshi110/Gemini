```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown
```

```python
def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
# use the gemini-pro-vision model and pass the image to the model with generate_content.
model = genai.GenerativeModel('gemini-pro-vision')
```

```python
import PIL.Image

img = PIL.Image.open('/content/drive/MyDrive/gemini/img gemini/young.png')
img
```

```python
# here itgives both text and image then generate the text.
response = model.generate_content(["Write a short, engaging blog post based on this picture. It should include a description of the photo.", img], stream=True)
response.resolve()
```

```python
to_markdown(response.text)
```
